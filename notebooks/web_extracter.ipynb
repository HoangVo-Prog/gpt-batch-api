```python
import pandas as pd
import trafilatura
from tqdm import tqdm

tqdm.pandas()     # kích hoạt tqdm cho pandas

# Đọc file CSV
df = pd.read_csv('gpt-batch-api/output/google_search.csv')

# Hàm lấy nội dung một URL
def extract_one(url):
    try:
        downloaded = trafilatura.fetch_url(url)
        if downloaded:
            return trafilatura.extract(downloaded)
        return ""
    except:
        return ""

# Hàm xử lý cột urls có dạng "url1|url2|url3..."
def split_and_extract(url_string):
    if pd.isna(url_string):
        return ["", "", "", "", ""]
    
    urls = [u.strip() for u in url_string.split("|")]
    urls = urls[:5]                  
    urls += [""] * (5 - len(urls))   

    contents = []
    for u in urls:
        if u == "":
            contents.append("")
        else:
            contents.append(extract_one(u))
    return contents

# Tạo 5 cột mới với tqdm theo dõi
df[["url1_text", "url2_text", "url3_text", "url4_text", "url5_text"]] = \
    df["urls"].progress_apply(split_and_extract).apply(pd.Series)

# Lưu kết quả
output_path = 'gpt-batch-api/output/web_extracted.csv'
df.to_csv(output_path, index=False)

output_path
```
